```python
boston=pd.read_csv('boston-housing/train.csv')
```

```python
print(boston.head())
headers=boston.dtypes.index
header=headers.tolist()
features=[boston['rm'],boston['lstat']]
Y=np.array(boston['medv'])
```

```python
X=[]
depth=len(Y)
width=len(features)
for i in range(depth):
    X.append([1])
    for j in range(width):
        X[i].append(features[j][i])
    #X[i].append(Y[i])
```

```python
X
```

```python
X=np.matrix(X)
Y=np.mat(Y).T
print(X)
#XT_X=np.dot(X.T,X)
#XT_Y=np.dot(X.T,Y)
```

```python
X.T
```

```python
inverse=np.dot(X.T,X).I
intermediate_rep=np.dot(inverse,X.T)
beta=np.dot(intermediate_rep,Y)
```

```python
beta
```

```python
new_y=np.dot(X,beta)
```

```python
#print(new_y)
error=Y-new_y
print(error)
```

```python
correct=0
for val in error:
    if val<=5:
        correct+=1
accuracy=correct/float(len(error))
print(accuracy)
```
